- 쇼핑몰 특정 페이지의 물품 리스트 가져오기 
- 물품명 목록과 가격을 합할것

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd 
pd.set_option('display.max_rows', 10000)

res = requests.get("https://www.musinsa.com/ranking/best?u_cat_cd=002")
soup = BeautifulSoup (res.content,'html.parser')

name = []
price = []

product = soup.find_all('p','list_info')
for item in product:
    name.append(item.get_text().strip())

product
```

```python
item_price = soup.find_all('p','price')
for item in item_price:
    price.append(item.get_text().strip())
```

```python
a = {'품목' : name ,'가격(판매가\할인가)' : price }
df = pd.DataFrame.from_dict(a, orient='index')
df.T
```
